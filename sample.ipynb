```python
def create_table():
    conn=psycopg2.connect("dbname='db1' user='postgres' password='postgre123' host='localhost' port='5432'")
    cur=conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS EMPLOYEE (EMPNAME TEXT, SALARY REAL, GRADE INT)")
    conn.commit()
    conn.close()
    
def insert(empname,salary,grade):
    conn=psycopg2.connect("dbname='db1' user='postgres' password='postgres123' host='localhost' port='5432'")
    cur=conn.cursor()
    cur.execute("Insert into employee values(%s,%s,%s)", (empname,salary,grade))
    conn.commit()
    conn.close()
    
def delete(empname=None,salary=None,grade=None):
    conn=psycopg2.connect("dbname='db1' user='postgres' password='postgres123' host='localhost' port='5432'")
    cur=conn.cursor()
    cur.execute("delete from employee where empname=%s or salary=%s or grade=%s", (empname,salary,grade))
    conn.commit()
    conn.close()
    
def view():
    conn=psycopg2.connect("dbname='db1' user='postgres' password='postgres123' host='localhost' port='5432'")
    cur=conn.cursor()
    cur.execute("select * from employee")
    rows=cur.fetchall()
    conn.close()
    return rows

def update(empname=None,salary=None,grade=None,empname1=None,salary1=None,grade1=None):
    conn=psycopg2.connect("dbname='db1' user='postgres' password='postgres123' host='localhost' port='5432'")
    cur=conn.cursor()
    cur.execute("update employee set empname=%s, salary=%s, grade=%s where empname=%s or salary=%s or grade=%s", (empname,salary,grade,empname1,salary1,grade1))
    conn.commit()
    conn.close()

                
create_table()
#insert('Bhupesh',26000,6)
#delete('',25000,)
#delete('',None,5)
update('Amit',25000,8,'Amit',None,None)
view()
```
